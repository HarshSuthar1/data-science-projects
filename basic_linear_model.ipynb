```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([2,3,4,5,6]).reshape(-1,1)
y = np.array([2,4,5,8,9])
```

```python
model = LinearRegression() 
model.fit(x,y)
```

```python
predict_y = np.array([8]).reshape(-1,1)

model.predict(predict_y)
```

```python
plt.scatter(x,y)
plt.plot(x,model.predict(x))
```

```python
from scipy.stats import norm

def calculate_normal_cdf():
    try:
        mean = float(input("Enter the mean (μ) of the normal distribution: "))
        std_dev = float(input("Enter the standard deviation (σ): "))
        x = float(input("Enter the value x to calculate P(X ≤ x): "))

        if std_dev <= 0:
            print("Standard deviation must be positive.")
            return

        # Calculate CDF
        probability = norm.cdf(x, loc=mean, scale=std_dev)
        print(f"P(X ≤ {x}) for N({mean}, {std_dev}²) = {probability:.4f}")

    except ValueError:
        print("Invalid input. Please enter numeric values.")

if __name__ == "__main__":
    calculate_normal_cdf()
```
