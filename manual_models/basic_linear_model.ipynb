```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
 
# x = [32,34,34,30,23,40,33,36]
x = np.array([32,34,34,30,23,40,33,36])
x.sort()
y = np.array([9,7,5,6,8,7,6,7])
y.sort()
plt.plot(x,y)
```

```python
a = (  (y.sum()*(x*x).sum()) - (x.sum()*(x*y).sum())  ) / (  (len(x)*(x*x).sum()) - (x.sum()*x.sum())  )
b = (  (len(x)*(x*y).sum()) - (x.sum()*y.sum())  ) / (  (len(x)*((x*x).sum())) -  (x.sum()*x.sum())  )
print(a)
print(b)
```

```python
def predict(alpha,beta,x_i):
    return beta * x_i +alpha
```

```python
def error(alpha,beta,x_i,y_i):
    return y_i - predict(alpha,beta,x_i) 

def sum_of_squared_errors(alpha,beta,x,y):
    sum=0
    for x_i,y_i in zip(x,y):
        sum = sum + error(alpha,beta,x_i,y_i) ** 2
    return sum
```

```python
# predict(a,b,31)

sum_of_squared_errors(a,b,x,y)
```
