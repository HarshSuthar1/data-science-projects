```python
import pandas as pd 
from matplotlib import pyplot as plt
second = pd.Series([2,3,4,5,6,7,8,9])
mark = pd.Series([0,1,0,1,0,1,0,1])
third = pd.Series([2,4,5,3,7,9,6,10]).sort_values()

# plt.figure(figsize=(7,7))
# plt.subplots_adjust(hspace=0.4,wspace=0.4)

plt.subplot(2,2,1)
plt.plot(first,second,marker='s',linestyle='solid',color='blue',label='blue')
plt.subplot(2,2,2)
plt.plot(first,mark,marker='^',linestyle='dashed',color='orange',label='orange')
plt.subplot(2,2,3)
plt.plot(third,mark,marker='o',linestyle='dashed',color='cyan',label='orange')
plt.subplot(2,2,4)
plt.plot(second,third,linestyle='dashed',color='orange',label='pink')
# plt.legend(["blue","orange"])
# plt.legend()
plt.title("Graph for data")
plt.xlabel("First Data")
plt.ylabel("Second Data")
# plt.grid()
# plt.tight_layout()
plt.show()
```
