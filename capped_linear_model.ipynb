```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Student_Marks.csv")
print(df.shape)
df.head()
```

```python
x = np.array(df.time_study).reshape(-1,1)
y = np.array(df.Marks)
```

```python
model = LinearRegression()
model.fit(x,y)
```

```python
study = np.array([17]).reshape(-1,1)
preds = np.clip(model.predict(study),a_min=0,a_max=100)
preds
```

```python
plt.plot(x,model.predict(x))
plt.scatter(x,y)
```
