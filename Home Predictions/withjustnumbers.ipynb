```python
train = pd.read_csv("train.csv",index_col="Id")
test = pd.read_csv("test.csv",index_col="Id")
# train.head()
test.head()
```

```python
X = train.copy()
X.pop("SalePrice")
X.head()
```

```python
# train.isnull().sum()
X.isnull().sum()
```

```python
y = train["SalePrice"]
# test_y = test["SalePrice"]
y
```

```python
string_columns = [i for i in X.columns if X[i].dtype == "object"]
```

```python
X = X.drop(string_columns,axis=1)
```

```python
train_X,val_X,train_y,val_y = train_test_split(X,y,random_state=1,train_size=0.85)
```

```python
model = RandomForestRegressor(n_estimators=100)

model.fit(train_X,train_y)
```
