```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
from sklearn.ensemble import RandomForestRegressor
import seaborn as sns
from sklearn.metrics import mean_squared_error
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import SGDRegressor
```

```python
data = pd.read_csv('winequality-red.csv')
data.head()
```

```python
data.shape
```

```python
# data.isnull().sum()
```

```python
data.info()
```

```python
X_data = data.drop(columns='quality')
y_data = data.quality
```

```python
train_X,test_X,train_y,test_y = train_test_split(X_data,y_data,train_size=0.78)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
train_X = scaler.fit_transform(train_X)
test_X = scaler.transform(test_X)
```

```python
model = SGDRegressor(
    max_iter=1000,       # More iterations
    tol=1e-3,            # Convergence tolerance
    penalty='l2',        # Regularization
    alpha=0.0001,        # Regularization strength
    learning_rate='invscaling',
    eta0=0.01,
    random_state=42
)
model.fit(train_X,train_y)
```

```python
preds = model.predict(test_X)
```

```python
mean_squared_error(preds,test_y)
```

```python
accuracy_score(test_y,preds)
```

```python
model.score(test_X,test_y)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
    'max_features': ['auto', 'sqrt']
}

grid = GridSearchCV(RandomForestRegressor(random_state=42), param_grid, cv=5, scoring='r2')
grid.fit(train_X, train_y)

print("Best R² Score:", grid.best_score_)
print("Best Params:", grid.best_params_)
```

```python
from sklearn.ensemble import GradientBoostingRegressor

gb = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, random_state=42)
gb.fit(train_X, train_y)
print("Gradient Boosting R² Score:", gb.score(test_X, test_y))
```
