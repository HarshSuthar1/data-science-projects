```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

```python
data = pd.read_csv("heart_disease_health_indicators_BRFSS2015.csv")
data.head()
data = data.sort_values(by='HeartDiseaseorAttack')
```

```python
data.columns
```

```python
# data.Smoker.value_counts()
```

```python
X_data = data.drop(columns=["HeartDiseaseorAttack"])
```

```python
y_data = data.HeartDiseaseorAttack
```

```python
X = np.array(X_data.BMI).reshape(-1,1)
```

```python
model = LogisticRegression()
model.fit(X,y_data)
```

```python
preds = model.predict(X)
```

```python
plt.plot(X_data.BMI,preds)
```
